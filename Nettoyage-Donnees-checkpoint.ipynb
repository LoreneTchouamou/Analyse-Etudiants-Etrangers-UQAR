```python
import re

# 1) Fonction pour uniformiser les libellés bruts du fichier
def norm(q):
    # strip + réduire espaces internes
    q2 = re.sub(r"\s+", " ", str(q).strip())
    return q2

# 2) Construire la table de correspondance (après normalisation)
rename_map_raw = {
    "Horodateur": "date_reponse",
    "Quel est votre pays d’origine ?": "pays_origine",
    "À quel campus de l'UQAR êtes-vous principalement rattaché(e) ?": "campus",
    "À quel cycle d’études êtes-vous inscrit(e) ?": "cycle_etudes",
    "Dans quel domaine étudiez-vous ?": "domaine_etudes",
    "Année d’entrée à l’UQAR  :": "annee_entree_uqar",
    "Avez-vous déjà échoué un ou plusieurs cours depuis votre arrivée à l'UQAR ?": "echec_cours",
    "Avez-vous déjà envisagé d’abandonner vos études à l'UQAR ?": "abandon_envisage",
    "Quelle est votre moyenne cumulative approximative actuelle ?": "moyenne",
    "Pensez-vous que votre formation antérieure vous a bien préparé(e) aux exigences académiques de l’UQAR ?": "preparation_formation",
    "Avant votre arrivée à l’UQAR, diriez-vous que vous aviez une bonne compréhension du programme d’études dans lequel vous vous êtes inscrit(e) ?": "comprehension_programme",
    "Avez-vous reçu une bourse d’études (d'un organisme, gouvernement, UQAR, etc.) ?": "bourse",
    "Travaillez-vous pendant vos études ?": "travail",
    "Si oui, combien d’heures par semaine travaillez-vous en moyenne ?": "nombre_heure",
    "Quelles sont les principales difficultés que vous avez rencontrées depuis votre arrivée à l'UQAR ? (Vous pouvez choisir plusieurs options)": "difficultes_principales",
    "Quelle session (ou période) a été la plus difficile pour vous à l’UQAR jusqu'à présent ?": "session_difficile",
    "Pourquoi cette session a-t-elle été difficile ?": "raison_session_difficile",
    "Diriez-vous que vous avez réussi à créer un réseau social solide (amis, connaissances) depuis votre arrivée à l'UQAR ?": "reseau_social_solide",
    "Interagissez-vous régulièrement avec des étudiants/personnes québécoises ou canadiennes ?": "contact_etudiants_quebec",
    "Connaissez-vous les services d'aide spécifiques offerts aux étudiants étrangers par l'UQAR (ex: soutien à l'intégration, aide linguistique, services d'aide psychologique) ?": "connaissance_services_uqar",
    "Avez-vous déjà utilisé les services d'aide offerts par l'UQAR (académiques, personnels, financiers, etc.) ?": "recours_service_aide",
    "Dans l'ensemble, votre expérience à l'UQAR et au Québec correspond-elle à vos attentes initiales ?": "experience_correspond_attentes",
    "Souhaitez-vous ajouter un commentaire ou partager un aspect de votre expérience d’étudiant(e) étranger(e) à l’UQAR ?": "commentaires_suggestions",
}

# 3) Normaliser les noms des colonnes actuelles et fabriquer le mapping réel
curr_cols = list(data.columns)
norm_to_actual = {norm(c): c for c in curr_cols}  # "propre" -> nom exact du fichier
rename_map = {}

for k_raw, v_new in rename_map_raw.items():
    k_norm = norm(k_raw)
    if k_norm in norm_to_actual:
        rename_map[norm_to_actual[k_norm]] = v_new  # map: libellé exact du fichier -> nom court

# 4) Appliquer le renommage
data = data.rename(columns=rename_map)

# 5) Contrôles et diagnostic
print("Colonnes renommées ->", len(rename_map))
print("Exemples:", dict(list(rename_map.items())[:5]))

missing_expected = [k for k in rename_map_raw.keys() if norm(k) not in norm_to_actual]
if missing_expected:
    print("⚠️ Libellés attendus non trouvés (orthographe/espaces différents ?) :")
    for m in missing_expected:
        print(" -", m)

print("\nAperçu des colonnes finales :")
print(list(data.columns))
```

```python
# Dictionnaire des corrections (écris en minuscules, sans accents ni espaces superflus)
corrections_ci = {
    "camerounaise": "Cameroun",
    "camerounais": "Cameroun",
    "republique du congo": "Congo",
    "congo cg": "Congo",
    "cote d ivoire": "Côte d'Ivoire",
    "maroc": "Maroc",
    "senegal": "Sénégal",
    "guinee": "Guinée",
    "guineenne": "Guinée"
}

# Fonction de nettoyage : normalise le texte et applique la correction
def corriger_pays(val):
    if pd.isna(val):
        return val
    txt = str(val).strip().lower()
    return corrections_ci.get(txt, val)  # si trouvé dans dict → remplace, sinon garde tel quel

# Appliquer à la colonne pays_origine
data["pays_origine"] = data["pays_origine"].apply(corriger_pays)

# Vérifier le résultat
print(data["pays_origine"].value_counts(dropna=False))
```

```python
def parse_moyenne(txt):
    if pd.isna(txt):
        return pd.Series([np.nan, np.nan, np.nan, "Inconnu"], index=["min","max","mid","bucket"])
    
    t = str(txt).replace(",", ".").strip()
    low = t.lower()

    # Cas "je ne sais pas"
    if "ne sais pas" in low or "préfère" in low:
        return pd.Series([np.nan, np.nan, np.nan, "Inconnu"], index=["min","max","mid","bucket"])

    # Retirer (sur 4.3) ou autres parenthèses
    t = re.sub(r"\(.*?\)", "", t).strip()
    nums = re.findall(r"\d+(?:\.\d+)?", t)

    if "entre" in low and len(nums) >= 2:
        a, b = float(nums[0]), float(nums[1])
        return pd.Series([a, b, (a+b)/2, f"{a}-{b}"], index=["min","max","mid","bucket"])
    if "supérieure" in low and nums:
        a = float(nums[0]); return pd.Series([a, np.nan, a, f">={a}"], index=["min","max","mid","bucket"])
    if "inférieure" in low and nums:
        b = float(nums[0]); return pd.Series([np.nan, b, b, f"<{b}"], index=["min","max","mid","bucket"])
    if len(nums) == 1:
        x = float(nums[0]); return pd.Series([x, x, x, f"{x}"], index=["min","max","mid","bucket"])
    return pd.Series([np.nan, np.nan, np.nan, "Inconnu"], index=["min","max","mid","bucket"])

# Appliquer
out = data["moyenne"].apply(parse_moyenne)
data[["moyenne_min","moyenne_max","moyenne_mid","moyenne_bucket"]] = out
data["moyenne_mid_pct"] = (data["moyenne_mid"] / 4.3) * 100

# Vérifier
print(data[["moyenne","moyenne_min","moyenne_mid","moyenne_bucket"]].head(10))
```

```python
def parse_heures(txt):
    if pd.isna(txt):
        return pd.Series([np.nan, np.nan, np.nan], index=["min","max","mid"])
    
    t = str(txt).lower().replace("heures","").replace("heure","").strip()

    # Cas intervalle "21-30"
    m = re.match(r"(\d+)\s*-\s*(\d+)", t)
    if m:
        a, b = float(m.group(1)), float(m.group(2))
        return pd.Series([a, b, (a+b)/2], index=["min","max","mid"])

    # Cas valeur unique "10"
    if t.isdigit():
        x = float(t)
        return pd.Series([x, x, x], index=["min","max","mid"])

    # Sinon : invalide
    return pd.Series([np.nan, np.nan, np.nan], index=["min","max","mid"])

# Appliquer sur la colonne
out = data["nombre_heure"].apply(parse_heures)

# Ajouter les colonnes numériques
data["nb_heure_min"] = out["min"]
data["nb_heure_max"] = out["max"]
data["nb_heure_mid"] = out["mid"]

# Vérifier
print(data[["nombre_heure","nb_heure_min","nb_heure_max","nb_heure_mid"]].head(10))
```

```python
# ========= Helpers =========
def crosstab_plot(df, row_col, outcome_col="echec_cours", title="", top_k=None):
    """
    Crosstab proportions par modalité (ligne) + barres empilées.
    - row_col: variable explicative (catégorielle)
    - outcome_col: variable de résultat (Oui/Non)
    - top_k: garder seulement les top_k modalités les plus fréquentes (utile pour 'domaine_etudes' ou 'pays')
    """
    s = df[row_col]
    if top_k is not None:
        top_vals = s.value_counts().head(top_k).index
        df_ = df[df[row_col].isin(top_vals)].copy()
    else:
        df_ = df.copy()

    tab = pd.crosstab(df_[row_col], df_[outcome_col], normalize="index").round(2)
    print(f"\n[{row_col}] → proportions par '{outcome_col}'")
    print(tab)

    ax = tab.plot(kind="bar", stacked=True)
    ax.set_title(title if title else f"{outcome_col} selon {row_col}")
    ax.set_xlabel(row_col); ax.set_ylabel("Proportion")
    ax.legend(title=outcome_col)
    plt.tight_layout(); plt.show()

# Tranches d'heures (si pas déjà créé)
if "nb_heure_mid" in data.columns:
    tranches_heures = pd.cut(
        data["nb_heure_mid"],
        bins=[-0.1,10,20,30,100],
        labels=["0–10h","11–20h","21–30h",">30h"]
    )
    data["heures_tranches"] = tranches_heures

# ========= 1) Échec × Heures =========
if {"heures_tranches","echec_cours"}.issubset(data.columns):
    crosstab_plot(
        data, "heures_tranches", "echec_cours",
        title="Échec × Heures de travail rémunéré"
    )

# ========= 2) Échec × Campus =========
if {"campus","echec_cours"}.issubset(data.columns):
    crosstab_plot(
        data, "campus", "echec_cours",
        title="Échec × Campus"
    )

# ========= 3) Échec × Expérience perçue =========
if {"experience_correspond_attentes","echec_cours"}.issubset(data.columns):
    crosstab_plot(
        data, "experience_correspond_attentes", "echec_cours",
        title="Échec × Expérience perçue"
    )

# ========= 4) Échec × Session difficile =========
if {"session_difficile","echec_cours"}.issubset(data.columns):
    crosstab_plot(
        data, "session_difficile", "echec_cours",
        title="Échec × Session jugée la plus difficile"
    )

# ========= 5) Échec × Cycle d’études =========
if {"cycle_etudes","echec_cours"}.issubset(data.columns):
    crosstab_plot(
        data, "cycle_etudes", "echec_cours",
        title="Échec × Cycle d’études"
    )

    
# Harmonisation des domaines d'études (exemples)
data["domaine_etudes"] = (
    data["domaine_etudes"]
    .str.strip()
    .str.lower()
    .replace({
        "baccalaureat en informatique": "informatique",
        "informatique": "informatique",
        "admin": "administration",
        "administration des affaires": "administration",
        "administration": "administration"
    })
    .str.capitalize()
)

# ========= 6) Échec × Domaine d’études (Top 8) =========
if {"domaine_etudes","echec_cours"}.issubset(data.columns):
    crosstab_plot(
        data, "domaine_etudes", "echec_cours",
        title="Échec × Domaine d’études (top 6)", top_k=6
    )

# ========= 7) Échec × Réseau social solide =========
if {"reseau_social_solide","echec_cours"}.issubset(data.columns):
    crosstab_plot(
        data, "reseau_social_solide", "echec_cours",
        title="Échec × Réseau social solide"
    )

# ========= 8) Échec × Connaissance des services UQAR =========
if {"connaissance_services_uqar","echec_cours"}.issubset(data.columns):
    crosstab_plot(
        data, "connaissance_services_uqar", "echec_cours",
        title="Échec × Connaissance des services UQAR"
    )

# ========= 9) Échec × Bourse =========
if {"bourse","echec_cours"}.issubset(data.columns):
    crosstab_plot(
        data, "bourse", "echec_cours",
        title="Échec × Bourse"
    )

# ========= 10) Échec × Travail (Oui/Non) =========
if {"travail","echec_cours"}.issubset(data.columns):
    crosstab_plot(
        data, "travail", "echec_cours",
        title="Échec × Travail (Oui/Non)"
    )

print("\n✅ Pack facteurs de risque exécuté.")
```

```python
# --- paramètres à adapter si besoin ---
INPUT_XLSX = r"C:\Users\loren\Documents\Documents\Projet_uqar\formulaire_etudiants_v1.xlsx"
SHEET_NAME = "Reponses_nettoyees"          # feuille des données
TEXT_COL   = "commentaires_suggestions"     # colonne texte
OUT_PNG    = r"C:\Users\loren\Documents\nuage_mots_commentaires.png"
OUT_FREQ   = r"C:\Users\loren\Documents\top_mots_commentaires.csv"

# --- imports / install si nécessaire ---
import sys, subprocess
def pip_install(pkg):
    try:
        __import__(pkg)
    except ModuleNotFoundError:
        subprocess.check_call([sys.executable, "-m", "pip", "install", pkg])

pip_install("pandas")
pip_install("wordcloud")
pip_install("unidecode")
pip_install("matplotlib")

import pandas as pd
from wordcloud import WordCloud, STOPWORDS
from unidecode import unidecode
import matplotlib.pyplot as plt
from collections import Counter
import re
import os

# 1) Charger les données
df = pd.read_excel(INPUT_XLSX, sheet_name=SHEET_NAME)

# 2) Récupérer le texte (on ignore NaN)
texts = df[TEXT_COL].dropna().astype(str)

# 3) Nettoyage léger : minuscules, retirer accents, garder lettres/apostrophes
def normalize_text(s: str) -> str:
    s = s.lower()
    s = unidecode(s)                     # retire accents (é->e) pour homogénéiser
    s = re.sub(r"[^a-z'\s-]", " ", s)    # garde lettres, espaces, ' et -
    s = re.sub(r"\s+", " ", s).strip()
    return s

cleaned = [normalize_text(t) for t in texts]

# 4) Stopwords FR (tu peux en ajouter/retirer)
stop_fr = {
    "et","ou","de","des","du","la","le","les","un","une","en","dans","pour","par","au","aux",
    "sur","avec","sans","pas","ne","que","qui","quoi","quand","ou","donc","car","mais",
    "je","tu","il","elle","on","nous","vous","ils","elles","me","moi","toi","leur","leurs",
    "ce","cet","cette","ces","mon","ma","mes","ton","ta","tes","son","sa","ses","notre","nos","votre","vos",
    "tres","plus","moins","bien","mal","a","deja","encore","aussi","cela","ca","comme","afin","entre",
    "uqar","etudiants","etudiant","etudes","etudier"  # spécifiques à ton contexte (peut être utile de garder/retirer)
}

# 5) Tokenisation + filtrage stopwords / mots courts
tokens = []
for t in cleaned:
    words = t.split()
    words = [w for w in words if w not in stop_fr and len(w) > 2]
    tokens.extend(words)

# 6) Fréquences (pour info et export CSV)
freq = Counter(tokens)
pd.DataFrame(freq.most_common(50), columns=["mot","frequence"]).to_csv(OUT_FREQ, index=False, encoding="utf-8-sig")

# 7) Générer le nuage de mots
#    (font_path aide à bien afficher les accents si tu ne faisais pas unidecode)
font_path = r"C:\Windows\Fonts\arial.ttf" if os.path.exists(r"C:\Windows\Fonts\arial.ttf") else None

wc = WordCloud(
    width=1600, height=900,
    background_color="white",
    stopwords=STOPWORDS.union(stop_fr),
    max_words=200,
    collocations=False,      # évite de coller des bigrammes répétitifs
    font_path=font_path
).generate(" ".join(tokens))

# 8) Afficher + sauvegarder
plt.figure(figsize=(12,7))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.tight_layout()
plt.show()

wc.to_file(OUT_PNG)
print("✅ Nuage de mots enregistré :", OUT_PNG)
print("✅ Top fréquences enregistré :", OUT_FREQ)
```
